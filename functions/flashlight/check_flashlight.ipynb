## Import libraries and define paths

```python
#--------------------------------------------------------------------------
# Import libraries
#--------------------------------------------------------------------------

import  os
import  shutil
import  glob
import  cv2

from    tqdm        import  tqdm
```

```python
#--------------------------------------------------------------------------
# Import libraries
#--------------------------------------------------------------------------

INPUT_DIR       =   '../../data/input/check_flashlight'
OUTPUT_DIR      =   '../../data/output/check_flashlight'

video_path_lst  =   glob.glob(os.path.join(INPUT_DIR, '*.mp4'))

#--------------------------------------------------------------------------
# Clean the output
#--------------------------------------------------------------------------

if  os.path.exists(OUTPUT_DIR):
    shutil.rmtree(OUTPUT_DIR)
os.mkdir(OUTPUT_DIR)
```

## Videos2Frames

```python
#--------------------------------------------------------------------------
# Video2Frames
#--------------------------------------------------------------------------
print("="*75)
print("= Processing Video2Frame")
for video_path in tqdm(video_path_lst):
    
    save_dir    =   os.path.join(OUTPUT_DIR, os.path.basename(video_path).split('.')[0], 'raw')
    os          .makedirs(save_dir, exist_ok=True)
    
    video_cap   =   cv2.VideoCapture(video_path)
    ret, frame  =   video_cap.read()
    count       =   0
    
    
    while ret:
        frame_path  =   os.path.join(save_dir, 'frame_{}.jpg'.format(str(count).zfill(5)))
        cv2         .imwrite(frame_path, frame)
        ret, frame  =   video_cap.read()
        count       +=  1
    
#--------------------------------------------------------------------------
# All Frames path
#--------------------------------------------------------------------------
frame_path_lst      =   glob.glob(os.path.join(OUTPUT_DIR, '*', 'raw', '*jpg'))
```
